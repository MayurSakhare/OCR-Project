# Spelling Checker

```python
from spellchecker import SpellChecker
 
spell = SpellChecker()
 
# find those words that may be misspelled
misspelled = spell.unknown(["THE QUICK BROWN FOX JUMPED OVER THE LAZY DOG."," the quick brown fox jumped over the lazy dog."," 01235456769"])
 
for word in misspelled:
    # Get the one `most likely` answer
    print(spell.correction(word))
 
    # Get a list of `likely` options
    print(spell.candidates(word))
```

```python
import pytesseract
from pytesseract import Output
import cv2
import matplotlib.pyplot as plt
from IPython.display import Image
from pylab import rcParams
import numpy as np
from pprint import pprint
from textblob import TextBlob

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

rcParams['figure.figsize'] = 8,16

file_name = 'E:\images\computer.jpg'
img = cv2.imread(file_name)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))


text_out = pytesseract.image_to_string(img)
print(text_out)
print(" ")
print("***********************************************************************************************************************")
print(" ")

a = text_out          # incorrect spelling
print("original text: "+str(a))
print(" ")
print("***********************************************************************************************************************")
print(" ")
 
b = TextBlob(a)
 
# prints the corrected spelling
print("corrected text: "+str(b.correct()))

print(" ")
print("***********************************************************************************************************************")
print(" ")

import nltk
nltk.download('punkt')
from nltk import word_tokenize,sent_tokenize
sentence_data = text_out
nltk_tokens = nltk.sent_tokenize(sentence_data)
print ("Sentenceswise Tokenizarion --- ",nltk_tokens)

print(" ")

# sentence_data = "THE QUICK BROWN FOX JUMPED OVER THE LAZY DOG. the quick brown fox jumped over the lazy dog. 01235456769 "
nltk_tokens1 = nltk.word_tokenize(sentence_data)
print ("Wordwise Tokenizarion ---", nltk_tokens1)
```
